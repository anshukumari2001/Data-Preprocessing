```python
from sklearn import preprocessing
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv("age_salary.csv",index_col=0)
data.head(5)
```

# Min Max Scalar

```python
x_real= data.iloc[:,[1,2]]
```

## formula= Xi-Xmin/Xmax-Xmin 

```python
minmaxscalar= preprocessing.MinMaxScaler(feature_range=(0, 1))
feature_scale=minmaxscalar.fit_transform(x_real)
```

```python
feature_scale
```

# Standard Scaler

```python
x_real= data.iloc[:,[1,2]]
```

# # formula= Xi-Xmean/(standard deviation of that feature)

```python
standardscaler= preprocessing.StandardScaler()
feature_scale=standardscaler.fit_transform(x_real)
```

```python
feature_scale
```

# Normalizer

```python
x_real= data.iloc[:,[1,2]]
```

# #formula= Xi/square of all the value of that row

```python
normalizer= preprocessing.Normalizer()
feature_scale=normalizer.fit_transform(x_real)
```

```python
feature_scale
```

# Binarizer

```python
binarizer= preprocessing.Binarizer(21)
feature_scale=binarizer.fit_transform(x_real)
```

```python
feature_scale
```
